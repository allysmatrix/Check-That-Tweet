### Types of sources within DBSCAN Cluster -1

**Observations**: The Cluster -1 has the highest target score among all clusters, which indicates the highest potential for tweets to be impactful. The Cluster -1 is marked by high follower count and thus high impact capability. The Cluster -1 contains a lot of outliers = users with large number of followers, public figure, traditional news sources like CNN, WSJ, AP, USAToday etc.  

We expect the twitter accounts in this cluster to have a wide reach and a much higher likelyhood of their tweets to be reteweeted or responded to. The mean retweet_count for this cluster is 27.33 vs the mean of 4.98 for all clusters.
